```python
import tkinter as tk
from tkinter import messagebox

class TicTacToe:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Tic-Tac-Toe Game")
        self.window.geometry("300x350")
        self.window.configure(bg="lightblue")
        self.player_turn = "X"

        self.create_buttons()
        self.create_control_buttons()

    def create_buttons(self):
        self.buttons = []
        for i in range(3):
            row = []
            for j in range(3):
                button = tk.Button(self.window, text="", command=lambda row=i, column=j: self.click(row, column), height=3, width=6)
                button.grid(row=i, column=j, padx=5, pady=5)
                row.append(button)
            self.buttons.append(row)

    def create_control_buttons(self):
        restart_button = tk.Button(self.window, text="Restart", command=self.restart_game, height=2, width=10)
        restart_button.grid(row=3, column=0, columnspan=2, pady=10)

        close_button = tk.Button(self.window, text="Close", command=self.window.quit, height=2, width=10)
        close_button.grid(row=3, column=2, columnspan=1, pady=10)

    def click(self, row, column):
        if self.buttons[row][column]['text'] == "":
            self.buttons[row][column]['text'] = self.player_turn
            if self.check_win():
                self.window.title("Player " + self.player_turn + " wins!")
                messagebox.showinfo("Game Over", "Player " + self.player_turn + " wins!")
                self.disable_buttons()
            else:
                self.player_turn = "O" if self.player_turn == "X" else "X"

    def check_win(self):
        for row in self.buttons:
            if row[0]['text'] == row[1]['text'] == row[2]['text']!= "":
                return True
        for column in range(3):
            if self.buttons[0][column]['text'] == self.buttons[1][column]['text'] == self.buttons[2][column]['text']!= "":
                return True
        if self.buttons[0][0]['text'] == self.buttons[1][1]['text'] == self.buttons[2][2]['text']!= "":
            return True
        if self.buttons[0][2]['text'] == self.buttons[1][1]['text'] == self.buttons[2][0]['text']!= "":
            return True
        return False

    def disable_buttons(self):
        for row in self.buttons:
            for button in row:
                button['state'] = 'disabled'

    def restart_game(self):
        for row in self.buttons:
            for button in row:
                button['text'] = ""
                button['state'] = 'normal'
        self.player_turn = "X"
        self.window.title("Tic-Tac-Toe Game")

    def run(self):
        self.window.mainloop()

if __name__ == "__main__":
    game = TicTacToe()
    game.run()
```
